```python
from analyse.counter import Counter
from collections import Counter as CollCount


fake_sites = ["fake/coronanuchterheid_v2.json","fake/staopvoorvrijheid-articles.json", "fake/stichtingvaccinvrij-articles.json", "fake/transitieweb-articles.json"]
counters = []
for site in fake_sites:
    c = Counter(real=False)
    c.add_sites([site])
    counters.append(c)

real_sites =['real/nos.json']

c = Counter(real=True)
c.add_sites(real_sites)
counters.append(c)
# Zodra er meer echte sites zijn toegevoegd, vervang de 3 regels hierboven met hetgene wat hierna komt:
# for site in real_sites:
#     c = Counter(real=True)
#     c.add_sites(site)
#     counters.append(c)
```

```python
def find_bigrams(counter, t=0):
    fdist = counter.get_bigrams()
    keys = set()
    for k,v in fdist.items():
        if v>t:
            keys.add(k)
    return keys

def shared_bigrams(counter_a, counter_b, t):
    """ Outputs bigrams which appear in both sets, at least more times then t"""
    keys_a = find_bigrams(counter_a,t)
    keys_b = find_bigrams(counter_b,t)
    return keys_a & keys_b

def shared_bigrams_multiple_sets(counters, nr_sites, appearences, real):
    """ Outputs bigrams that appear at least t times and at least in x sets with the same label for real"""
    all_bigrams = set()
    all_sets = []
    final_set = set()
    for counter in counters:
        if counter.real==real:
            new_set = find_bigrams(counter,appearences)
            all_sets.append(new_set)
            all_bigrams = all_bigrams|new_set
    for bigram in all_bigrams:
        count = 0
        for i in all_sets:
            if bigram in i:
                count+=1
        if count>=nr_sites:
            final_set.add(bigram)
    return final_set


def different_bigrams(counters,nr_fake_sites,nr_real_sites,appearences):
    """ Outputs words which appear in at least x fake sites at least t times, but not in z real (below threshold)"""
    frequent_fake_bigrams = shared_bigrams_multiple_sets(counters,nr_fake_sites,appearences,False)
    frequent_real_bigrams = shared_bigrams_multiple_sets(counters,nr_real_sites,appearences,True)
    return frequent_fake_bigrams^frequent_real_bigrams
```

```python
def find_words(counter, appearences=0):
    fdist = counter.get_words()
    keys = set()
    for k,v in fdist.items():
        if v>appearences:
            keys.add(k)
    return keys

def shared_words(counter_a, counter_b, appearences):
    """ outputs a set of words that appear at least t times in both a and b"""
    keys_a = find_words(counter_a,appearences)
    keys_b = find_words(counter_b,appearences)
    return keys_a & keys_b

def shared_words_multiple_sets(counters, nr_sites, appearences, real):
    """ Outputs words that appear at least t times and at least in x sets with the same label for real"""
    all_words = set()
    all_sets = []
    final_set = set()
    for counter in counters:
        if counter.real==real:
            new_set = find_words(counter,appearences)
            all_sets.append(new_set)
            all_words = all_words|new_set
    for word in all_words:
        count = 0
        for i in all_sets:
            if word in i:
                count+=1
        if count>=nr_sites:
            final_set.add(word)
    return final_set


def different_words(counters,nr_fake_sites,nr_real_sites,appearences):
    """ Outputs words which appear in at least nr_fakes_sites fake sites at least t times, but not in nr_real_sites real (below threshold)"""
    frequent_fake_words = shared_words_multiple_sets(counters,nr_fake_sites,appearences,False)
    frequent_real_words = shared_words_multiple_sets(counters,nr_real_sites,appearences,True)
    return frequent_fake_words^frequent_real_words
```
